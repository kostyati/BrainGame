```python
from tkinter import*
from random import choice
from tkinter import messagebox


#global variables
char_list=[chr(char) for char in range(97,123)]
level=1
wait_time=2000

root=Tk()
root.title('Memo Game')
root.geometry('400x400')
#root.iconbitmap('')

#Start game window
def start_game():
      
    def show_label():
        start_butt.destroy()
        
        def check(): 
            answer=ans_entry.get()
    
            ans_label=Label(root, text='Your answer is '+answer)
            ans_label.grid(row=3,column=3)
    
            if answer==question:
                resp=messagebox.askyesno(f'You are correct!','Do you wanna play enother game?')    
                if resp==1:
                    
                    global level
                    level+=1
                    global wait_time
                    wait_time+=500
                    
                    new_label.destroy()
                    check_butt.destroy()
                    ans_label.destroy()
                    ans_entry.destroy()
                    start_game()
                else:
                    root.destroy()
                    
            else:
                resp=messagebox.askyesno(f'Sorry.You are wrong.','Do you wanna play another game?')  
                if resp==1:
            
                    new_label.destroy()
                    check_butt.destroy()
                    ans_label.destroy()
                    ans_entry.destroy()
                    start_game()
                else:
                    root.destroy()
                
        
        def clear_label():
            mem_label.destroy()
            
            global new_label
            new_label=Label(root, text='Enter Your Answer: ')
            new_label.grid(row=0, column=0) 
    
            global ans_entry
            ans_entry=Entry(root)
            ans_entry.grid(row=0, column=1)
            
            global check_butt
            check_butt=Button(root, text='Check!', command=check)
            check_butt.grid(row=1, column=1)
     
        def rand_text():
            global rand_string
            rand_string=''
            for num in range(level):
                rand_string+=choice(char_list)
            return rand_string
        
        question=rand_text()
        mem_label=Label(root, text=question, font=('Helvetica',30))
        
        global wait_time
        mem_label.place(x=150,y=150)
        root.after(wait_time,clear_label)
        
    global level
    start_butt=Button(root, text=f'Start level {level}',command=show_label)
    start_butt.pack(anchor=CENTER)
    

start_game()
root.mainloop()
```
